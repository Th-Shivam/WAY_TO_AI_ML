## How do i rename columns in a dataframe 

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
ufo.head()
```

```python
ufo.rename(columns = {"Colors Reported":"Colors_Reported" , "Shape Reported":"Shape_Reported"})
```

```python
ufo_cols = ['city' , 'colors reported' , 'shape reported' , 'state' , 'time']
ufo.columns = ufo_cols
```

```python
ufo.head()
```

```python
ufo.columns = ufo.columns.str.replace(' ' , '_')
```

```python
ufo.columns
```

## How do i remove columns from a pandas dataframe ?

```python
ufo1 = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo1.drop("Colors Reported" , axis = 1 , inplace = True)
```

```python
ufo1.head()
```

```python
ufo1.drop(["City" , "State"] , axis = 1 , inplace = True)
ufo1.head()
```

```python
# Droping rows with .drop 
ufo1.drop([0,1] , axis = 0)
```

## How do i sort a pandas dataframe or series ?

```python
movies1 = pd.read_csv("http://bit.ly/imdbratings")
movies1.head()
```

```python
# Newer version to sort the values 
movies1.title.sort_values()
```

```python
# Sorting the values in descending order 
movies1.title.sort_values(ascending = False)
```

```python
movies1.sort_values('duration')
```

```python
movies1.head()
```

```python
movies1.sort_values(['content_rating','duration'])
```
