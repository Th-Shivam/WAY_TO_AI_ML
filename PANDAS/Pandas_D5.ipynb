```python
import pandas as pd 
```

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies.head()
```

```python
movies[movies.duration>=200]
```

```python
# Filtering the dataframe based on two condition and also with logical "&" and Logical or "|" 
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]
```

```python
# if we want to filter our dataframe on the basis of multiple where conditions then , us .isin() method

movies[movies.genre.isin(['Crime','Action'])]
```
