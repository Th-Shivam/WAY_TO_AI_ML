```python
import json 


def load_json_data():
    with open("codebook_data.json", "r") as f:
        data = json.load(f)
        return data
```

```python
data = load_json_data()
```

```python
for user in data["users"]:
    print(f"{user['name']} (ID: {user['id']}) - Friends: {user['friends']} - Liked Pages: {user['liked_pages']}")
```
