```python
for i in range(1,51) :
    htmlfile = rq.get(f"https://books.toscrape.com/catalogue/page-{i}.html")

    if(htmlfile.status_code == 200):
        with open(f"Htmls/page-{i}.html" , "w") as f :
            f.write(htmlfile.text) 
            print(f"Page {i} scrapped and saved succesfully .")
    else:
        print(f"Error occured , server returned {htmlfile.status_code}")
        
    
print("Program executed successfully .")    
    
```
