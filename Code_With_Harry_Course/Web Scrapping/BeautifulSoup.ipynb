```python
from bs4 import BeautifulSoup as BS
import requests as rq
```

```python
response = rq.get("https://books.toscrape.com/catalogue/page-1.html")
soup = BS(response.text , "html.parser")
```

```python
soup.find_all("h3")[0].find("a").text
```

```python
books = soup.find_all("h3")
for book in books:
    title = book.find("a")["title"]
    print(title)
```

```python
articles = soup.find_all("article")
```

```python
for article in articles:
    h3s = article.find("h3")
    print(h3s.text)
```

```python
for article in articles:
    ratings = article.find("p")
    print(ratings)
```

```python
for article in articles:
    price = article.find("p", class_="price_color")
    print(price)

```
