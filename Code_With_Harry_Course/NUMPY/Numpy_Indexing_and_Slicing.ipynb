# Indexing and slicing

```python
import numpy as np 
```

```python
arr = np.array([10, 20, 30, 40])
print(arr[0])  # 10
print(arr[-1]) # 40
```

# Slicing (Extracting Parts of an Array)

```python
arr1 = arr[1:3]
print(arr1)
```

## Slicing returns a view, not a copy! Changes affect the original array.

```python
arr2 = arr[1:3].copy()
arr2
```

# Fancy indexing

```python
arr = np.array([10, 20, 30, 40])
indexes = [2,3]
arr[indexes]
```

# Boolean masking

```python
arr = np.array([10, 20, 30, 40])
mask = arr>25
arr[mask]
```
