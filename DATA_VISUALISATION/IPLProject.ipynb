```python
# Which team won how many matches 
match_wins = df['match_winner'].value_counts()
sns.barplot(x = match_wins.values , y = match_wins.index , palette="rainbow")
```

```python
sns.countplot(x = 'toss_decision' , data = df )
```

```python
t_and_m_winner = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = ( t_and_m_winner / df.shape[0])*100
print(percentage)
```
