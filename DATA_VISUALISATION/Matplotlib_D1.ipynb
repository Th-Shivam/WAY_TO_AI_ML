```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,10,5)
y = x**2
```

```python
# Simple line graph
plt.plot(x,y)
plt.title("Hello to the world of DATA SCIENCE")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
# Subplot in the canvas
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x , x )
plt.subplot(2,2,4)
plt.plot(x , y**2)
```

```python
# To work on a specific plot of the canvas 
fig = plt.figure(figsize=[12,8] , dpi=500)
#  the params here are => (left, bottom, width, height)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.4,0.3,0.3])
axis2.plot(x,y)
plt.title("SHIVAM")
plt.savefig("fig2.png")
```

Now in the above plot , we have the flexibilty to control the plot figure as per our requirement .
