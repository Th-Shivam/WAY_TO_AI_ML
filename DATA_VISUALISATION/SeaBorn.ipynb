# SEABORN DATA VISUALISTAION

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'] , kde = 'true')
plt.subplot(1,2,2)
sns.histplot(df['tip'] , kde = 'true')
```

```python
sns.jointplot(x='total_bill' , y  = 'tip' , data = df , kind = 'scatter')
```

```python
sns.jointplot(x='total_bill' , y  = 'tip' , data = df , kind = 'kde')
```

```python
sns.jointplot(x='total_bill' , y  = 'tip' , data = df , kind = 'hex')
```

```python
sns.jointplot(x='total_bill' , y  = 'tip' , data = df , kind = 'reg')
```

```python
sns.jointplot(x='total_bill' , y  = 'tip' , data = df , kind = 'resid')
```

```python
sns.pairplot(df , hue="sex")
```

# Categorical data visualisation 

```python
sns.countplot(x = df['sex'] , hue = df['smoker'])
```

```python
import numpy as np
sns.barplot(x=df['sex'] , y = df['tip'] , estimator=np.sum , hue=df['smoker'])
```

```python
sns.violinplot(x='day' , y = 'total_bill' , data = df , palette="rainbow")
```

```python
sns.barplot(x='tip' , y = 'day' , data = df)
```

```python
sns.swarmplot(x='tip' , y='day' , data=df)
sns.violinplot(x='tip' , y='day' , data=df)
```
