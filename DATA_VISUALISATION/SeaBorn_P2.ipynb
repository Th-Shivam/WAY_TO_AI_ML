```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr = tips[['total_bill' , 'tip' , 'size']]
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr() , annot=True)
```

```python
sns.clustermap(tipscorr.corr() , annot=True)
```

```python
pvtflight = flights.pivot(index='month' , columns='year' , values='passengers')
pvtflight
```

```python
sns.heatmap(pvtflight)
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'sex' , markers='v')
```
